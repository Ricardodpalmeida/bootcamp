## Visualization

### Numerical Variables

```python
#Correlations
hr_corr = hr_ds.corr(method = 'spearman')
hr_corr
```

```python
#Heat map 1
hr_corr = hr_ds.corr(method = 'spearman')
figure = plt.figure(figsize = (16,10))
sns.heatmap(hr_corr, annot = True, fmt = '.1g')
#sns.heatmap(hr_corr)
```

```python
#Heat Map 2: coloured, without annotations
hr_corr = hr_ds.corr(method = 'spearman')
figure = plt.figure(figsize = (16,16))
sns.heatmap(hr_corr,  cmap = 'coolwarm', annot = True, fmt = '.1g', vmin = -1, vmax = 1)
plt.show()
```

```python
#Heat Map 4: coloured: Red to Blue , with annotations and with min and max values
hr_corr = hr_ds.corr(method = 'spearman')
figure = plt.figure(figsize = (16,16))
sns.heatmap(hr_corr,  cmap = 'RdBu',  annot = True, square = True, fmt = '.1g')
plt.show()
```

```python
hr_corr_p = hr_ds.corr(method = 'pearson')
figure = plt.figure(figsize = (16,16))
sns.heatmap(hr_corr_p,  cmap = 'RdBu',  annot = True, square = True, fmt = '.1g')
plt.show()
```
