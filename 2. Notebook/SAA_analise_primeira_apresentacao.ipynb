```python
cross_tab_bt = pd.crosstab(index=df['BusinessTravel'],
                             columns=df['Attrition'],
                             normalize="index")
cross_tab_bt
```

```python
cross_tab_bt.plot(kind='bar', 
                    stacked=True, 
                    color=['#bed62f', '#5c666c'], 
                    figsize=(10, 6),
                    legend=False)

#plt.legend(loc="upper left", ncol=2)
plt.xlabel("BusinessTravel")
plt.ylabel("Attrition")

for n, x in enumerate([*cross_tab_bt.index.values]):
    for (proportion, y_loc) in zip(cross_tab_bt.loc[x],
                                   cross_tab_bt.loc[x].cumsum()):
                
        plt.text(x=n - 0.17,
                 y=(y_loc - proportion) + (proportion / 2),
                 s=f'{np.round(proportion * 100, 1)}%', 
                 color="black",
                 fontsize=12,
                 fontweight="bold")
plt.show()
plt.savefig('businesstravel_attrition.png')
```

Attrition seems to be higher in employees that travel frequently
